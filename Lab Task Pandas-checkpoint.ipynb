```python
#- Count Total Number of cars for each company
df = pd.read_csv("D:\data.csv")
df['make'].value_counts()
```

```python

#- Print all BMW car details
df = pd.read_csv("D:\data.csv")
car_Manufacturers = df.groupby('make')
bmwDf = car_Manufacturers.get_group('bmw')
bmwDf
```

```python

# Sort all cars by num-of-cylinders

cars = pd.read_csv("D:\data.csv")
cars = cars.sort_values(by=['num-of-cylinders'], ascending=False)
cars.head(1000)
```

```python

#Find average car price of each company 

df = pd.read_csv("D:\data.csv")
car = df.groupby('make')
price = car['make','price'].mean()
price
```

```python
#cheapest cars of all companies
df = pd.read_csv("D:\data.csv")
car= df.groupby('make')
price = car['make','price'].min()
price
```

```python
#identify the most cheapest car   
df = pd.read_csv("D:\data.csv")  
df.nsmallest(1, ['price'])
```
